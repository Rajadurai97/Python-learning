# Sets

Sets in Python are unordered collections of unique elements. They are mutable, meaning you can add or remove elements after the set is created. Sets are denoted by curly braces {}, similar to dictionaries, but without the key-value pairs.

```python
my_set = set()
```

```python
type(my_set)
```

```python
my_set.add(1)
my_set.add(2)
```

```python
my_set
```

```python
s = {1,2,3,4,1} # set sytored only a unique values
```

```python
s
```

```python
s.add(5)
```

```python
s
```

```python
my_list = [1,1,11,1,22,2,2,2,4,45,5]
```
